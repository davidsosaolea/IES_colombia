```python
# Funcion para eliminar las filas que no contienen informacion
def eliminar_filas_iniciales(df, n):
    df = df.iloc[n:].reset_index(drop=True)  # Eliminar las primeras n filas y restablecer los índices
    df.columns = df.iloc[0]  # Establecer la nueva primera fila como encabezado
    df = df[1:].reset_index(drop=True)  # Eliminar la antigua primera fila y restablecer los índices nuevamente
    return df
```

```python
# muestra los valores unicos por columna
def obtener_resumen_valores_unicos(df):
    # Obtener el número de valores únicos en cada columna
    valores_unicos = df.nunique()

    # Obtener el número total de valores en cada columna
    total_valores = df.count()

    # Combinar los resultados en un nuevo DataFrame
    resumen_valores_unicos = pd.DataFrame({
        'Columna': valores_unicos.index,
        'Valores Únicos': valores_unicos.values,
        'Total Valores': total_valores.values
    })
    
    return resumen_valores_unicos

# Imprimir el resumen de valores únicos
print(obtener_resumen_valores_unicos(df))
```

```python
# muestra los valores nulos 
# Obtener la cantidad de valores 
valores_nulos = df.isnull().sum()

print(valores_nulos)
```

```python
# Obtenenos los valores unicos y comparamos las colunas con ID vs los nombres
def obtener_diccionario(df, columna_a, columna_b, n = 2):
    '''
    Crea un diccionario que relaciona los códigos únicos de la columna_a con los nombres únicos de la columna_b.
    
    Parámetros:
        - df: DataFrame, el DataFrame de donde se obtendrán los datos.
        - columna_a: str, nombre de la columna que contiene los códigos.
        - columna_b: str, nombre de la columna que contiene los nombres.
        - n: int, valor mínimo de nombres únicos requeridos para incluirlos en el diccionario (por defecto, n=2).
        
    Retorna:
        Un diccionario donde las claves son los códigos únicos y los valores son listas de nombres únicos 
        asociados a cada código, siempre y cuando haya al menos n nombres únicos para ese código.
    '''   
    codigos_unicos = df[columna_a].unique()
    diccionario = {}

    # Iterar sobre los códigos únicos
    for codigo in codigos_unicos:
        # Obtener los valores únicos de los nombres para cada código
        nombres_unicos = df[df[columna_a] == codigo][columna_b].unique()

        # Verificar si hay más de 2 valores únicos de nombres
        if len(nombres_unicos) >= n:
            # Asignar los nombres únicos como valor para cada código en el diccionario
            diccionario[codigo] = nombres_unicos.tolist()

    return diccionario
```

```python

resultado = obtener_diccionario(df, 'IES PADRE', 'Institución de Educación Superior (IES)',1)
resultado1 = obtener_diccionario(df, 'IES PADRE', 'Institución de Educación Superior (IES)')
resultado2 = obtener_diccionario(df, 'Institución de Educación Superior (IES)', 'IES PADRE')

# Imprimir la cantidad de códigos únicos con al menos 2 nombres distintos asociados.
print(len(resultado1))
# Imprimir el resultado2, que muestra los nombres únicos de cada código.
print(resultado2)

# Iterar sobre los valores del resultado2 (que son las listas de nombres únicos para cada código).
for i in resultado2.values():
    for n in i:
        if len(resultado[n]) >= 2:
            print('El codigo/ID {} tiene 2 categorias distintas {}'.format (n,resultado[n]))
```

```python
# Esta funcion entrega un diccionario con los valores a ser reemplazado 
def obtener_diccionario_remplaza(d_f, col_codigos, col_valores):
    # Obtenemos los códigos únicos en la columna 'col_codigos'
    codigos_unicos = d_f[col_codigos].unique()
    diccionario = {}
    
     # Creamos un diccionario donde las claves son los códigos únicos y los valores son listas de nombres únicos
    for codigo in codigos_unicos:
        nombres_unicos = d_f[df[col_codigos] == codigo][col_valores].unique()
        
        if len(nombres_unicos) > 1:
            diccionario[codigo] = nombres_unicos.tolist()
            
    # Creamos una lista con los primeros valores de cada lista de nombres y construimos un nuevo diccionario
    lista_valores = [valor[0] for valor in diccionario.values()]
    diccionario_actualizado = {lista_valores[i]: valor for i, valor in enumerate(diccionario.values())}
    
    # Construimos el diccionario invertido
    diccionario_invertido = {}

    for clave, valores in diccionario_actualizado.items():
        for valor in valores:
            diccionario_invertido.setdefault(valor, []).append(clave)
    
    return diccionario_invertido
```

```python
# Obtenemos los valores y remplazamos
p2 = obtener_diccionario_remplaza(df, 'Código de  la Institución', 'Institución de Educación Superior (IES)')
print(p2)
```

```python
# Reemplazamos los valores
p2 = {k: v[0] for k, v in p2.items()}
df["Institución de Educación Superior (IES)"] = df["Institución de Educación Superior (IES)"].replace(p2)
```
